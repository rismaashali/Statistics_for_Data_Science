```python
NA1 = titan[(titan['Pclass'] == 1) & (titan['Survived'] == 1)].shape[0]
NS = titan[(titan['Survived'] == 1)].shape[0]
NA2 = titan[(titan['Pclass'] == 2) & (titan['Survived'] == 1)].shape[0]
NA3 = titan[(titan['Pclass'] == 3) & (titan['Survived'] == 1)].shape[0]
```

```python
#probability class 1
Pel1 = NA1/NS
print(Pel1)
```

```python
#probability class 2
Pel2 = NA2/NS
print(Pel2)
```

```python
#probability class 3
Pel3 = NA3/NS
print(Pel3)
```
